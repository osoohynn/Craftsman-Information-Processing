```python
import pandas as pd

train_dataset = pd.read_csv('/content/drive/MyDrive/train-data.csv')
test_dataset = pd.read_csv('/content/drive/MyDrive/test-data.csv')

train_dataset.head()
```

```python
train_dataset.isnull().sum()
```

```python
train_dataset.drop(columns=['Unnamed: 0', 'Name', 'Location', 'New_Price'], inplace=True)
```

```python
test_dataset.drop(columns=['Unnamed: 0', 'Name', 'Location', 'New_Price'], inplace=True)
```

```python
train_dataset.head()
```

```python
kmkg = 0
kmpl = 0
for i in train_dataset.Mileage:
    if str(i).endswith("km/kg"):
        kmkg+=1
    elif str(i).endswith("kmpl"):
        kmpl+=1
print('The number of rows with Km/Kg : {} '.format(kmkg))
print('The number of rows with Km/Kg : {} '.format(kmpl))
```

```python
train_dataset = train_dataset.dropna(subset=['Mileage'])
```

```python
Correct_Mileage = []
for i in train_dataset.Mileage:
    if str(i).endswith('km/kg'):
        i = i[:-6]
        i = float(i)*1.40
        Correct_Mileage.append(float(i))
    elif str(i).endswith('kmpl'):
        i = i[:-6]
        #print(i)
        Correct_Mileage.append(float(i))

train_dataset['Mileage'] = Correct_Mileage
```

```python
train_dataset['Engine'] = train_dataset['Engine'].str.extract(r'(\d+)')
train_dataset.head()
```

```python
train_dataset['Power'] = train_dataset['Power'].str.extract(r'(\d+)')
train_dataset.head()
```

```python
train_dataset.isnull().sum()
```

```python
engine_mean = train_dataset['Engine'].dropna().astype(int).mean().astype(int)
train_dataset['Engine'].fillna(engine_mean, inplace=True)

print(engine_mean)
```

```python
power_mean = train_dataset['Power'].dropna().astype(int).mean().astype(int)
train_dataset['Power'].fillna(power_mean, inplace=True)

print(power_mean)
```

```python
seats_mean = train_dataset['Seats'].dropna().astype(int).mean().astype(int)
train_dataset['Seats'].fillna(seats_mean, inplace=True)

print(seats_mean)
```

```python
train_dataset.isnull().sum()
```

```python
train_dataset['Fuel_Type'].value_counts()
```

```python
train_dataset['Fuel_Type'].replace({'Petrol': 0, 'Diesel': 1, 'CNG': 2, 'LPG':3}, inplace=True)
train_dataset.head()
```

```python
train_dataset['Transmission'].value_counts()
```

```python
train_dataset['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
train_dataset.head()
```

```python
train_dataset['Owner_Type'].value_counts()
```

```python
train_dataset['Owner_Type'].replace({'First': 0, 'Second': 1, 'Third': 2, 'Fourth & Above':3}, inplace=True)
train_dataset.head()
```

```python
X = train_dataset.drop(columns=['Price'])
Y = train_dataset['Price']
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X, Y)

lr.score(X, Y)
```
